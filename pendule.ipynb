```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler(f, y0, dydt0, D, h):
    
    N = int(D/h) # nombre de valeurs a calculer
    
    t = np.zeros(N)
    y = np.zeros(N)
    dydt = np.zeros(N)
    
    t[0] = 0.0
    y[0] = y0
    dydt[0] = dydt0
    
    for i in range(N-1):
        
        t[i+1] = t[i] + h
        
        dydt[i+1] = dydt[i] + h * f( y[i], dydt[i], t[i] )
        
        y[i+1] = y[i] + h * dydt[i]
    
    return (t,y,dydt)
```

```python
g = 9.81 # m/s**2
R = 0.50 # m

def f(y, dydt, t):
    return -g/R * np.sin(y)
```

```python
theta0 = np.pi/18
t, theta, dtheta_dt = euler(f, theta0, 0.0, 10, 1.e-4)
```

```python
plt.plot(t,theta,'-',label="Euler")
plt.plot(t, theta0*np.cos( (g/R)**0.5 * t ), '-',label="Approximation harmonique")
plt.legend()
plt.show()
```
